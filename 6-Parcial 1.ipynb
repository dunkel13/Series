```python
### Leer base_Accidentes.xlsx
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
import scipy as sp
from statsmodels.tsa.arima_process import arma_acf
from statsmodels.tsa.arima_process import ArmaProcess
import statsmodels as sm
import statsmodels.api as smapi
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import acovf
from statsmodels.tsa.arima_model import ARIMA
```

```python
data2=pd.ExcelFile('/Users/usuario/Documents/Base_Accidentes.xlsx')
data2=data2.parse('DATA2005')
ind = pd.date_range(start='1/1/2005', end='8/1/2018', freq='M')
accidentes2 = data2.set_index(ind)
AUTP = pd.Series(accidentes2['AUTP'],index=ind)
AUTP= AUTP[:-1,]
plt.plot(AUTP)
acf(AUTP, nlags=50,unbiased=False)
ResAUTP=plot_acf(AUTP,lags=50,unbiased=False, alpha=0.05)
```

Gráficamente se ve que la serie no presenta una componente de tendencia, también se puede decir que oscila alrededor de un valor medio
